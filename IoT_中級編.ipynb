# Raspberry Piについて

## マイクロコンピュータ(マイコン)とRaspberryPiの違い

### Raspberry Piのほうが便利
- モニタ出力, LAN, WiFi, USB, Cameraなどのインタフェースの接続
- ハードウェア・リソースが大きい
- GUIに対応したOSが動作する

### マイコンのほうが便利

- リアルタイム性 (μs) ラズパイは(ms)
- 外部入出力 A-Dコンバータや高速パルスの入出力がない
- 立ち上がりが早い
- 電源断に対する配慮がない



# プログラミング概要

## プログラミングとは

コンピュータが実行するプログラムを作成する作業のことをプログラミングといいます。
プログラムにはコンピュータが処理する順番が記述されています。日本語や英語があるようにプログラムを記述する言葉も様々あり、この言葉のことをプログラミング言語と呼びます。今回取り上げるPythonという言語もこの中の一つです。その他、スマートフォン用のアプリを開発するための言語や、Webサイトを構築するための言語など、用途によって様々な特徴を持つ言語が存在しています。

コンピュータは高度なAIを除き、プログラムの途中で人間のように状況判断を行うことできないため、動作する順番を一つ一つ書いて命令する必要があります。しかし、プログラムを書いただけではコンピュータが解釈できる言葉ではなく、実行するためにはプログラムに書いている内容を解釈し、コンピュータが理解できる言葉へ変換する必要があります。

コンピュータが理解できる言葉へ変換するタイミングによって２つに分類できます。
- インタープリタ方式
- コンパイル方式

### インタープリタ方式

プログラムの内容を渡すとその場でコンピュータの言語に解釈し、実行してくれる方式をインタープリタ方式といいます。代表的な言語としてPythonやPHPなど比較的動作の軽い処理に利用されています。実行するとプログラムを解釈しながら実行を行うため、コンパイル方式に比べてすぐに実行できるという特徴があります。解釈を行いながら実行するため何度も同じプログラムを実行する場合には無駄が発生する可能性があります。

### コンパイル方式

プログラムの内容をコンピュータが解釈できる内容に変換し、機械だけが実行できる形式に保存して利用する方式をコンパイル方式といいます。代表的な言語としてC言語やJavaなどが挙げられます。一度機械語に解釈したプログラムは別途保存され何度も使うことができるため、インタープリタ方式のように都度変換の必要がなく無駄が発生せず何度も繰り返す処理や比較的重い動作を実行する際に利用されます。プログラムを変更した際には再度コンパイルが必要なため、頻繁に変更される動作などには向いておりません。
